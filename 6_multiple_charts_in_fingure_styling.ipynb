```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
from datetime import timedelta
import numpy as np
```

```python
# Create figure and axes
fig, ax = plt.subplots(2, 2, figsize=(12, 4), sharex = 'all')  # 1 row, 3 columns

# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# First subplot
ax[0][0].plot(x, y1, color='r')
ax[0][0].set_title("Sine Wave")

# Second subplot
ax[0][1].plot(x, y2, color='g')
ax[0][1].set_title("Cosine Wave")

# Third subplot
ax[1][0].plot(x, y3, color='b')
ax[1][0].set_title("Tangent Wave")
ax[1][0].set_ylim(-5, 5)  # Limit y-axis to avoid extreme values

ax[1][1].set_axis_off()

# Adjust layout
plt.tight_layout()
plt.show()
```
