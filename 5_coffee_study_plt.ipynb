```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
from datetime import timedelta
import numpy as np
```

```python
prod = pd.read_csv('coffee/total-production.csv', index_col = 'total_production')
prod.head()
```

```python
top_10_producers = prod.sum(axis = 'columns').sort_values(ascending = False)[:10]

figure = plt.figure()
ax = figure.add_subplot()

plt.xticks(rotation=45)
ax.set_title('Top 10 Coffee Producers (1990-2018)')
ax.set_xlabel('Countries')
ax.set_ylabel('Coffee Production (in million Kgs)')

ax.bar(top_10_producers.index, top_10_producers.values / 1000000)
```

```python
top_5_producers = top_10_producers.index[:5]
top_5_over_time = prod.loc[top_5_producers]
top_5_over_time
```

```python
top_5_producers = top_10_producers.index[:5]
top_5_over_time = prod.loc[top_5_producers]

figure = plt.figure(figsize = (9, 5))
ax = figure.add_subplot()

plt.xticks(rotation=45)
ax.set_title('Top 5 Coffee Producers Production over time (1990-2018)')
ax.set_xlabel('Years')
ax.set_ylabel('Coffee Production (in Kgs)')


for i in top_5_over_time.index:
    ax.plot(pd.to_datetime(top_5_over_time.loc[i].index), top_5_over_time.loc[i].values, label = i)

ax.set_xticks(pd.to_datetime(top_5_over_time.columns[::4]))
ax.xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y'))


ax.legend()
```

```python
brazil_data = prod.loc['Brazil']
rest_world_data = prod.drop(index = ['Brazil'])
rest_world_data = rest_world_data.sum(axis = 'index')

figure = plt.figure(figsize = (9, 5))
ax = figure.add_subplot()

ax.set_xticks(pd.to_datetime(brazil_data.index[::4]))
ax.xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y'))
ax.set_title("Brazil vs. Rest of the World's Coffee Production over time (1990-2018)")
ax.set_xlabel('Years')
ax.set_ylabel('Coffee Production (in 100,000 Kgs)')

ax.stackplot(brazil_data.index, brazil_data.values / 100000, rest_world_data.values / 100000, labels=['Brazil', 'Rest of the World'])
```

```python
brazil_data = prod.loc['Brazil']
VietNam_data = prod.loc['Viet Nam']

figure = plt.figure(figsize = (7, 5))
ax = figure.add_subplot()

ax.set_title("Brazil vs. Viet Nam's Coffee Production over time (1990-2018)")
ax.set_xlabel("Brazil's Production over the years (1990-2018) in Kgs")
ax.set_ylabel("Viet Nam's Production over the years (1990-2018) in Kgs")

ax.scatter(x = brazil_data, y = VietNam_data, color='blue', alpha=0.7)
```

```python
top_producers = prod.sum(axis = 'columns').sort_values(ascending = False)[:5].sort_values(ascending = False)
rest_world = prod.drop(index = top_producers.index).sum(axis = 'columns').sum(axis = 'index')
top_producers.loc['rest of the world'] = rest_world

figure = plt.figure()
ax = figure.add_subplot()

ax.set_title("Brazil vs. Viet Nam's Coffee Production over time (1990-2018)")

wedges, texts, autotexts = ax.pie(x = top_producers.values, labels = top_producers.index, startangle = 90, autopct = '%1.0f%%')
for i in autotexts:
    i.set_color('white')

plt.show()
```

```python
brazil_1990 = prod.loc['Brazil']['1990']
total_1990 = prod['1990'].sum()

progress = (brazil_1990 / total_1990) * 100

fig, ax = plt.subplots()

ax.pie([progress, 100 - progress], 
       colors=['blue', 'lightgray'], 
       startangle=90, 
       wedgeprops={'width': 0.3}, 
        counterclock=False
)

ax.text(0, 0, f"{round(progress)}%", ha='center', va='center', fontsize=35, color='black')

ax.set_aspect('equal')
ax.set_title("Brazil's Coffee Production Contribution in 1990")

plt.show()
```

```python
brazil_2018 = prod.loc['Brazil']['2018']
total_2018 = prod['2018'].sum()

progress = (brazil_2018 / total_2018) * 100

fig, ax = plt.subplots()

ax.pie([progress, 100 - progress], 
       colors=['blue', 'lightgray'], 
       startangle=90, 
       wedgeprops={'width': 0.3}, 
        counterclock=False
)

ax.text(0, 0, f"{round(progress)}%", ha='center', va='center', fontsize=35, color='black')

ax.set_aspect('equal')
ax.set_title("Brazil's Coffee Production Contribution in 2018")

plt.show()
```

```python
prices = pd.read_csv('coffee/retail-prices.csv')
prices['retail_prices'] = prices['retail_prices'].str.strip()
prices.set_index('retail_prices', inplace = True)

price_avg = prices.sum(axis = 'columns') / prices.shape[1]

imports = pd.read_csv('coffee/imports.csv')
imports['imports'] = imports['imports'].str.strip()
imports = imports.fillna(0)
imports.set_index('imports', inplace = True)
imports = imports.loc[price_avg.index]

import_avg = imports.sum(axis = 'columns') / imports.shape[1]
import_avg = import_avg.sort_values(ascending = False)
price_avg = price_avg.loc[import_avg.index]

prices = pd.read_csv('coffee/retail-prices.csv')
prices['retail_prices'] = prices['retail_prices'].str.strip()
prices.set_index('retail_prices', inplace = True)
prices = prices.loc[import_avg.index]

price_avg = prices.sum(axis = 'columns') / prices.shape[1]

x = np.arange(price_avg.shape[0])
width = 0.35

fig, ax = plt.subplots(figsize=(13, 7))
ax2 = ax.twinx()
ax.bar(x - width/2, import_avg.values, width = width, label="Imports", color="blue")
ax2.bar(x + width/2, price_avg.values, width = width, label="Prices", color="orange")


ax.set_title("Imports vs. Prices avg for countries", fontsize=14, fontweight='bold')
ax.set_xlabel("Countries", fontsize=12)
ax.set_ylabel("Import in Kgs", fontsize=12)
ax2.set_ylabel("Price in USD", fontsize=12)
ax.set_xticks(x)  # Set category positions
ax.set_xticklabels(import_avg.index, fontsize=10, rotation=45)  # Assign category names
fig.legend(loc="upper right", bbox_to_anchor=(0.9, 0.88))

plt.grid(axis='y', linestyle="--", alpha=0.6)  # Light y-axis grid for readability
```
