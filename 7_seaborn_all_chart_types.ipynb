```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
#  barchart

sns.barplot(x = 'day', y = 'total_bill', data = tips, estimator = np.mean, errorbar = ('ci', 95))  # default
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = tips, estimator = np.mean, errorbar = 'sd')
```

```python
sns.barplot(x = 'sex', y = 'tip', data = tips, hue = 'sex')
```

```python
#  horizontal bar chart

sns.barplot(x = 'tip', y = 'sex', data = tips)  # flip x and y params
```

```python
#  grouped bar chart

data = tips.sort_values(by = 'day')

sns.barplot(x = 'sex', y = 'total_bill', data = data, hue = 'day')  # use hue to differentiate days
```

```python
#  histogram

sns.histplot(x = 'total_bill', data = tips, bins = 30, kde = True)  # or use binwidth for bin width
```

```python
sns.histplot(x = 'total_bill', data = tips, bins = 30, kde = True, stat = 'percent')  # use percent as y
```

```python
sns.histplot(x = 'total_bill', data = tips, bins = 20, kde = True, hue = 'day')
```

```python
#  heatmap

sns.histplot(x = 'total_bill', y = 'tip', data = tips, cbar = True)
```

```python
#  scatterplot

sns.scatterplot(x = 'tip', y = 'total_bill', data = tips, hue = 'tip', palette = 'husl', size = 'size')
```

```python
sns.boxplot(x = 'total_bill', y = 'sex', data = tips)  # y is optional
plt.grid(True)
```

```python
#           min          (Q1  median  Q3)     Q3 + 1.5*IQR          max
#                     ( Inter Quartile Range)
```

```python
sns.violinplot(x = 'total_bill', y = 'sex', data = tips)
```

```python
tips
```

```python
#  regression plot

sns.regplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
# lm plot extension of regression plot

sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex')
```

```python
# lmplot matrix, use row and column

sns.lmplot(x = 'total_bill', y = 'tip', data = tips, row = 'day', col = 'sex', height = 3, aspect = 1)
```

```python
# jointplot default : kind = 'scatter'

sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips,kind = 'hist', bins = 30)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips,kind = 'kde', hue = 'sex', palette = 'viridis')
```

```python
# pairplot default

sns.pairplot(data = tips)
```

```python
sns.pairplot(data = tips, hue = 'sex', corner = True)
```

```python
#  heatmap

tips["bill_bin"] = pd.cut(tips["total_bill"], bins=range(0, int(tips["total_bill"].max()) + 5, 5))
temp = tips.pivot_table(index="bill_bin", columns="day", values="tip", aggfunc="mean", observed = True)
sns.heatmap(data = temp, annot = True)  # use annot to see values in boxes
```

```python
# facetgrid

g = sns.FacetGrid(tips, col = 'day', col_wrap = 2)
g.map_dataframe(sns.barplot, x = 'sex', y = 'tip')
```
