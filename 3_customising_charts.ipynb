```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
from datetime import timedelta
import numpy as np
```

```python
#  initialize df
data = [['2022-01-01', 10, 2], ['2022-02-01', 20, 4], ['2022-03-01', 30, 3], ['2022-04-01', 40, 1]]
df = pd.DataFrame(data, columns = ['Date', 'Sales', 'Profit'])
df["Date"] = pd.to_datetime(df["Date"])
df.set_index('Date', inplace = True)

#  initialize figure
fig = plt.figure()
ax = fig.add_subplot()

#  add title
fig.suptitle('Overall Title', fontsize = 14)
ax.set_title('Sub Title', fontsize = 12)
plt.xticks(rotation=45)  # rotate x axis labels

#  remove right and top box borders
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.plot(df.index, df['Sales'], label = 'India')
ax.plot(df.index, df['Profit'], ls = '--', label = 'China')

#  legend
ax.legend(['In', 'Ch'])  # optional change in legend otherwise uses label
ax.legend(['In', 'Ch'], loc = 'best')  # no change in legend pos
ax.legend(['In', 'Ch'], loc = 'lower right')
#  [
#    'best', 'upper right', 'upper left', 'lower left', 'lower right',
#    'right', 'center left', 'center right', 'lower center', 'upper center', 'center'
#  ]
ax.legend(['In', 'Ch'], bbox_to_anchor = (1, 1))  # 0 -> 1 on both axis
ax.legend(['In', 'Ch'], bbox_to_anchor = (1.2, 1))  # valid to send legend  outside limit
ax.legend(['In', 'Ch'], bbox_to_anchor = (1.1, 1), ncol = 2)  # number of columns to display legend horizontal
ax.legend(['In', 'Ch'], bbox_to_anchor = (1.1, 1), ncol = 2, frameon = False)  # boox frame around legend
```

```python
fig2 = plt.figure(figsize = (10, 5))  # use x / 2.54 for cm
ax2 = fig2.add_subplot()
fig2.suptitle('Big Title', y = 0.95)  #  adjust for figsize increase
ax2.set_xlim(18980, 19100)  # number of days after 1st Jan 1970
ax2.set_ylim(0, 50)

plt.xticks(rotation=45)  # rotate x axis labels

#  linestyles
ax2.plot(df.index, df['Sales'], color = 'green', linewidth = 5, linestyle = ':')  # '-', '--', or ':'
ax2.plot(df.index, df['Profit'], color = 'purple', linewidth = 3, linestyle = '--')
```

```python
#  change figure size
fig2 = plt.figure(figsize = (10, 5))  # use x / 2.54 for cm
ax2 = fig2.add_subplot()
fig2.suptitle('Big Title', y = 0.95)  #  adjust for figsize increase

# show specific x values
dates = ['2022-01-15', '2022-02-10', '2022-03-05']
dates = pd.to_datetime(dates)
ax2.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))  #  use mdates from matplotlib.dates module
ax2.set_xticks(dates)
plt.xticks(rotation=45)  # rotate x axis labels

ax2.plot(df.index, df['Sales'], color = 'green', linewidth = 5, linestyle = ':')  # '-', '--', or ':'
ax2.plot(df.index, df['Profit'], color = 'purple', linewidth = 3, linestyle = '--')
```

```python
fig2 = plt.figure()
ax2 = fig2.add_subplot()

plt.xticks(rotation=45)  # rotate x axis labels

#  vertical line
ax2.axvline(pd.to_datetime('2022-01-10'), color = 'grey', linestyle = '--', ymin = 0.05, ymax = 0.8, label = 'Important Date')

ax2.plot(df.index, df['Sales'])
ax2.plot(df.index, df['Profit'])

ax2.legend(loc = 'center right')
```

```python
fig2 = plt.figure()
ax2 = fig2.add_subplot()

plt.xticks(rotation=45)  # rotate x axis labels

#  vertical line
ax2.axvline(pd.to_datetime('2022-01-10'), color = 'grey', linestyle = '--', ymin = 0.05, ymax = 0.8)

#  text in ax anf figure
ax2.text(pd.to_datetime('2022-01-11'), 8, '<-- Important Date')
fig2.text(0.93, 0.8 , 'Extra\nInformation', fontname = 'Comic Sans MS', color = 'blue')

ax2.plot(df.index, df['Sales'])
ax2.plot(df.index, df['Profit'])
```

```python
fig = plt.figure()
ax = fig.add_subplot()

plt.xticks(rotation=45)  # rotate x axis labels

#  vertical line
ax.axvline(pd.to_datetime('2022-01-10'), color = 'grey', linestyle = '--', ymin = 0.05, ymax = 0.8)


#  annotation / annotate
ax.annotate('Important Date', xy = (pd.to_datetime('2022-01-11'), 8), xytext = (pd.to_datetime('2022-01-20'), 8),
           arrowprops = dict(facecolor = 'black', width = 0.7, headwidth = 6, connectionstyle = 'angle3, angleA=270, angleB=0'),
           verticalalignment = 'center')

ax.plot(df.index, df['Sales'])
ax.plot(df.index, df['Profit'])
```
