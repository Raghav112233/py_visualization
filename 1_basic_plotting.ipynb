```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2, 5, 6, 7, 8])  # basic plot using pyplot api
```

```python
xaxis = ['1', '2', '3', '4', '5']  # plot using pyplot api
yaxis = [10, 20, 30, 40, 50]
plt.plot(xaxis, yaxis)
plt.title('Title')
```

```python
fig = plt.figure()  # plot using object oriented
ax = fig.add_subplot()
xaxis = ['1', '2', '3', '4', '5']
yaxis = [10, 20, 30, 40, 50]
ax.plot(xaxis, yaxis)
fig.suptitle('Overall Title')
ax.set_title('Title')
```

```python
import pandas as pd
```

```python
data = [['2022-01-01', 10, 2], ['2022-02-01', 20, 4], ['2022-03-01', 30, 3], ['2022-04-01', 40, 1]]
df = pd.DataFrame(data, columns = ['Date', 'Sales', 'Profit'])
df.set_index('Date', inplace = True)
df
```

```python
df.plot()  # plot dataframe 
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(df.index, df['Sales'])
ax.plot(df.index, df['Profit'], linestyle = '--')
```
